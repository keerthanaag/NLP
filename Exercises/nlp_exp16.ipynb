```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Download the VADER lexicon
nltk.download('vader_lexicon')
```

```python
def load_whatsapp_chat(filepath):
    with open(filepath, 'r', encoding='utf-8') as file:
        chat_data = file.read()
    return chat_data
```

```python
def preprocess_chat_data(chat_data):
    processed_data = ''.join(char for char in chat_data if char.isalnum() or char.isspace())
    return processed_data
```

```python
def perform_sentiment_analysis(text):
    # Initialize VADER sentiment analyzer
    sid = SentimentIntensityAnalyzer()
    # Perform sentiment analysis
    sentiment_scores = sid.polarity_scores(text)
    # Return the compound score (overall sentiment score)
    return sentiment_scores['compound']
```

```python
def list_positive_words(text):
    # Tokenize the text into words
    words = text.split()
    # Initialize an empty list to store positive words
    positive_words = []
    # Initialize VADER sentiment analyzer
    sid = SentimentIntensityAnalyzer()
    # Iterate through each word and check if it has a positive sentiment
    for word in words:
        if sid.polarity_scores(word)['compound'] > 0:
            positive_words.append(word)
    return positive_words
```

```python
chat_filepath = 'sample_whatsapp.txt'

# Load WhatsApp chat data
chat_data = load_whatsapp_chat(chat_filepath)

# Preprocess chat data
preprocessed_data = preprocess_chat_data(chat_data)

# Perform sentiment analysis
chat_sentiment_score = perform_sentiment_analysis(preprocessed_data)
```

```python
if chat_sentiment_score >= 0.05:
    overall_sentiment = 'Positive'
elif chat_sentiment_score <= -0.05:
    overall_sentiment = 'Negative'
else:
    overall_sentiment = 'Neutral'

# List positive words in the chat
positive_words = list_positive_words(preprocessed_data)
print("Positive Words in the Chat:")
print(positive_words)

```
